```python
def label(flag: str) -> int:
    return {
        '-': 0,
        'snmpgetattack': 1,
        'warez': 2,
        'portsweep': 3,
        'nmap': 4,
        'warezclient': 5,
        'ipsweep': 6,
        'dict': 7,
        'neptune': 8,
        'smurf': 9,
        'pod': 10,
        'snmpguess': 11,
        'teardrop': 12,
        'satan': 13,
        'httptunnel-e': 14,
        'ignore': 15,
        'mscan': 16,
        'guest': 17,
        'rootkit': 18,
        'back': 19,
        'apache2': 20,
        'processtable': 21,
        'mailbomb': 22,
        'smurfttl': 23,
        'saint': 24
    }.get(flag)
```

```python
port_list = {key: [] for key in range(25)}

train_data_files = os.listdir('data/train/')
num_file = 0

for file in train_data_files:
    num_file += 1
    print('Loading File {}/{}...'.format(num_file, len(train_data_files)), end='\r')
    
    file_path = 'data/train/' + file
    train_data = open(file_path, 'r', encoding='latin-1')
    
    for line in train_data:
        source_ip, destination_ip, port, timestamp, connection_type = line.split()
        
        if label(connection_type) == 0:
            continue
        
        if int(port) in port_list[label(connection_type)]:
            continue
        else:
            port_list[label(connection_type)].append(int(port))
```
